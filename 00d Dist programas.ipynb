```python
ruta_labels = Programas_EFT['Ruta habilitada'].unique()

# Crear un diccionario con un DataFrame por cada valor de ruta
rutas_dict = {
    ruta: Programas_EFT.loc[Programas_EFT['Ruta habilitada'] == ruta].copy()
    for ruta in ruta_labels
}

# Ejemplos de acceso:
Antiguos = rutas_dict.get("Antiguos")
Nuevos = rutas_dict.get("Nuevos")
No_Habilitado = rutas_dict.get("NO HABILITADO")
Grupos_Cerrados = rutas_dict.get("Grupos cerrados")
```

```python
Antiguos['IO'] = Antiguos['NÚMERO DE CUPOS A OFERTAR'] * Antiguos['IPO'] 
Antiguos['IO'].describe()
```

```python
# 1. Calcular la suma total de IO
IO_total = Antiguos['IO'].sum()

# 2. Agrupar por cod_CNO y Ocupacion y agregar suma y conteo
IO_por_CNO = (
    Antiguos
    .groupby(['cod_CNO', 'Ocupacion'])
    .agg(
        IO=('IO', 'sum'),
        CUPOS=('NÚMERO DE CUPOS A OFERTAR', 'sum'),
        IPO=('IPO', 'sum'),
        n_programas=('IO', 'count')  # Cuenta los registros por grupo
    )
    .reset_index()
)

# 3. Calcular la participación de cada grupo sobre el total del IO
IO_por_CNO['IO_part'] = IO_por_CNO['IO'] / IO_total

# 4. Asignar recursos según la participación en el IO total
total_recursos = 1_320_000_000
IO_por_CNO['RecursosXCNO'] = IO_por_CNO['IO_part'] * total_recursos

# 5. Visualizar resultados
IO_por_CNO
```

```python
IO_por_CNO = IO_por_CNO[['cod_CNO', 'RecursosXCNO']]
```

```python
Antiguos = Antiguos.merge(
    IO_por_CNO[['cod_CNO', 'RecursosXCNO']],
    on='cod_CNO',
    how='left'
)
```

```python
Antiguos.head(1)
```

```python
# Paso 1: Ordenar programas por prioridad (dentro de cada ocupación)
Antiguos = Antiguos.sort_values(
    by=['cod_CNO', 'Ocupacion', 'IPO', 'ISOEFT_4d'],
    ascending=[True, True, False, False]
)

# Paso 2: Crear nueva columna para asignación
Antiguos['Cupos_Asignados_2E'] = 0

# Paso 3: Iterar por grupo de ocupación para asignar los recursos disponibles
for (cod_cno, ocupacion), grupo in Antiguos.groupby(['cod_CNO', 'Ocupacion']):
    recurso_total = grupo['RecursosXCNO'].iloc[0]  # mismo valor para todos en el grupo
    saldo = recurso_total
    indices = grupo.index

    for i in indices:
        costo_unitario = Antiguos.loc[i, 'VALOR DEL PROGRAMA SEGÚN DERECHOS PECUNIARIOS 2024 (SI APLICA)']
        cupos_disp = Antiguos.loc[i, 'NÚMERO DE CUPOS A OFERTAR']

        if pd.isna(costo_unitario) or costo_unitario == 0:
            continue  # saltar si no hay costo válido

        recurso_necesario = cupos_disp * costo_unitario

        if saldo >= recurso_necesario:
            Antiguos.loc[i, 'Cupos_Asignados_2E'] = cupos_disp
            saldo -= recurso_necesario
        else:
            cupos_asignables = saldo // costo_unitario
            Antiguos.loc[i, 'Cupos_Asignados_2E'] = cupos_asignables
            saldo -= cupos_asignables * costo_unitario
            break  # ya no hay más recursos en esta ocupación

# Paso 4: Calcular recursos efectivamente asignados
Antiguos['Recurso_Asignado_2E'] = Antiguos['Cupos_Asignados_2E'] * Antiguos['VALOR DEL PROGRAMA SEGÚN DERECHOS PECUNIARIOS 2024 (SI APLICA)']

# Paso 5: Sumar resultados
asignacion_por_ocupacion_alt = Antiguos.groupby(['cod_CNO', 'Ocupacion'])[['Recurso_Asignado_2E']].sum().reset_index()

# Paso 6: Obtener RecursosXCNO originales por ocupación
recursos_por_ocupacion = Antiguos.groupby(['cod_CNO', 'Ocupacion'])[['RecursosXCNO']].first().reset_index()

# Paso 7: Unir ambos DataFrames
asignacion_por_ocupacion_alt = asignacion_por_ocupacion_alt.merge(
    recursos_por_ocupacion, on=['cod_CNO', 'Ocupacion']
)

# Paso 8: Calcular el saldo no asignado
asignacion_por_ocupacion_alt['Saldo_No_Asignado_2E'] = (
    asignacion_por_ocupacion_alt['RecursosXCNO'] - asignacion_por_ocupacion_alt['Recurso_Asignado_2E']
)
```

```python
# Paso 9: Sumar todos los saldos no asignados en la segunda etapa
saldo_comun_3E = asignacion_por_ocupacion_alt['Saldo_No_Asignado_2E'].sum()

# Paso 10: Inicializar nueva columna para tercera etapa
Antiguos['Cupos_Asignados_3E'] = 0
Antiguos['Recurso_Asignado_3E'] = 0

# Paso 11: Ordenar programas por prioridad para tercera etapa
Antiguos_ordenados_3E = Antiguos.sort_values(
    by=['IPO', 'ISOEFT_4d'],
    ascending=[False, False]
)

# Paso 12: Asignación iterativa con la bolsa común
for idx, row in Antiguos_ordenados_3E.iterrows():
    costo = row['VALOR DEL PROGRAMA SEGÚN DERECHOS PECUNIARIOS 2024 (SI APLICA)']
    cupos = row['NÚMERO DE CUPOS A OFERTAR']

    if pd.isna(costo) or costo <= 0 or pd.isna(cupos) or cupos <= 0:
        continue

    recurso_necesario = costo * cupos

    if saldo_comun_3E >= recurso_necesario:
        # Financiar el programa completo
        Antiguos.at[idx, 'Cupos_Asignados_3E'] = cupos
        Antiguos.at[idx, 'Recurso_Asignado_3E'] = recurso_necesario
        saldo_comun_3E -= recurso_necesario
    else:
        # Ver si se puede financiar al menos un cupo
        cupos_posibles = saldo_comun_3E // costo
        if cupos_posibles >= 1:
            recurso_asignado = cupos_posibles * costo
            Antiguos.at[idx, 'Cupos_Asignados_3E'] = cupos_posibles
            Antiguos.at[idx, 'Recurso_Asignado_3E'] = recurso_asignado
            saldo_comun_3E -= recurso_asignado
        else:
            break  # No se puede financiar ni un cupo más
            
# Paso 12+1: Calcular totales asignados (segunda + tercera etapa)
Antiguos['Total_Cupos_Asignados'] = Antiguos['Cupos_Asignados_2E'] + Antiguos['Cupos_Asignados_3E']
Antiguos['Total_Recurso_Asignado'] = Antiguos['Recurso_Asignado_2E'] + Antiguos['Recurso_Asignado_3E']

# Paso 14: Agregar columna con el saldo restante total luego de la tercera etapa
Antiguos['Saldo_Remanente_3E'] = saldo_comun_3E
```

```python
# Paso 1: Ordenar programas por prioridad
Grupos_Cerrados = Grupos_Cerrados.sort_values(
    by=['Puntaje (nuevos y cerrados)', 'NÚMERO DE CUPOS A OFERTAR'],
    ascending=[False, False]
)

# Paso 2: Marcar empates
Grupos_Cerrados['Empate_Puntaje_Cupos'] = Grupos_Cerrados.duplicated(
    subset=['Puntaje (nuevos y cerrados)', 'NÚMERO DE CUPOS A OFERTAR'], keep=False
)

# Paso 3: Inicializar columnas
Grupos_Cerrados['Cupos_Asignados_2E'] = 0
Grupos_Cerrados['Recurso_Asignado_2E'] = 0
Grupos_Cerrados['Saldo_Total_Remanente'] = 0

# Paso 4: Inicializar saldo total
saldo_total = 990_000_000

# Paso 5: Asignar recursos fila a fila
for idx, row in Grupos_Cerrados.iterrows():
    costo = row['VALOR DEL PROGRAMA SEGÚN DERECHOS PECUNIARIOS 2024 (SI APLICA)']
    cupos = row['NÚMERO DE CUPOS A OFERTAR']

    if pd.isna(costo) or costo <= 0 or pd.isna(cupos) or cupos <= 0:
        Grupos_Cerrados.at[idx, 'Saldo_Total_Remanente'] = saldo_total
        continue

    recurso_necesario = costo * cupos

    if saldo_total >= recurso_necesario:
        Grupos_Cerrados.at[idx, 'Cupos_Asignados_2E'] = cupos
        Grupos_Cerrados.at[idx, 'Recurso_Asignado_2E'] = recurso_necesario
        saldo_total -= recurso_necesario
    else:
        cupos_posibles = saldo_total // costo
        recurso_asignado = cupos_posibles * costo
        Grupos_Cerrados.at[idx, 'Cupos_Asignados_2E'] = cupos_posibles
        Grupos_Cerrados.at[idx, 'Recurso_Asignado_2E'] = recurso_asignado
        saldo_total -= recurso_asignado

    # Registrar el saldo restante después de esta fila
    Grupos_Cerrados.at[idx, 'Saldo_Total_Remanente'] = saldo_total

    if saldo_total <= 0:
        break  # se agotaron los recursos
```
